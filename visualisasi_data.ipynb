```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv("customers-data.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset['Profession'].unique()
```

```python
prof = 'Lawyer'
dataset[dataset['Profession'] == prof]
```

```python
unique_gender = dataset[dataset.Work_Experience>0]['Gender'].unique()
unique_gender.sort()
unique_gender
```

```python
gender_workExp_case = []
for i in unique_gender:
  gender_workExp_case.append(dataset[dataset.Work_Experience>0][dataset['Gender'] == i][dataset['Profession'] == prof].Work_Experience.sum())
```

```python
print(set(zip(unique_gender, gender_workExp_case)))
```

```python
unique_age = dataset[dataset.Work_Experience>0]['Age'].unique()
unique_age.sort()
unique_age
```

```python
age_workExp_case = []
for i in unique_age:
  age_workExp_case.append(dataset[dataset.Work_Experience>0][dataset['Age'] == i][dataset['Profession'] == prof].Work_Experience.sum())
```

```python
print(set(zip(unique_age, age_workExp_case)))
```

```python
nan_indices = []
for i in range(len(unique_age)):
  if type(unique_age[i]) == float:
    nan_indices.append(i)
```

```python
unique_age = list(unique_age)
age_workExp_case = list(age_workExp_case)
```

```python
# menghilangkan nan/data kosong
for i in nan_indices: 
  unique_age.pop(i)
  age_workExp_case.pop(i)

unique_age
```

```python
plt.barh(unique_age, age_workExp_case)
plt.show()
```

```python
plt.figure(figsize=(50,40))
plt.barh(unique_age, age_workExp_case)
plt.show()
```

```python
plt.figure(figsize=(20,20))
plt.pie(age_workExp_case)
plt.legend(unique_age, loc='best')
plt.show()
```
