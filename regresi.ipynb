```python
# Import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
# Memanggil Dataset
dataset = pd.read_csv('Dataset_spine.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1:2].values
```

```python
#print(x)
print(y)
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
# Menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Split Dataset menjadi Training Set dan Testing Set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
# Melakukan Fitting Simple Linear Regression pada Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Memprediksi hasil Test-Set
y_pred = regressor.predict(x_test)
```

```python
# Visualisasi Data
plt.scatter(dataku.pelvic_incidence, dataku.pelvic_tilt)
plt.xlabel("Pelvic Incidence")
plt.ylabel("Pelvic Tilt")
plt.title("Grafik Pelvic Incidence vs Pelvic Tilt")
plt.show
```

```python
# Visualisasi Hasil Prediksi Pada Training-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(x_train[:,0], y_train, color = 'blue')

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color = 'red')

# Memberi Judul dan Label
plt.title('Pelvic Incidence terhadap Pelvic Tilt (Training Set)')
plt.xlabel('Pelvic Incidence')
plt.ylabel('Pelvic Tilt')

# Menampilkan plot
plt.show()
```

```python
# Visualisasi Hasil Prediksi Pada Test-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(x_test[:,0], y_test, color = 'blue')

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_test, regressor.predict(x_test), color = 'red')

# Memberi Judul dan Label
plt.title('Pelvic Incidence terhadap Pelvic Tilt (Test Set)')
plt.xlabel('Pelvic Incidence')
plt.ylabel('Pelvic Tilt')

# Menampilkan plot
plt.show()
```
