```python
# import library
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
dataset = pd.read_csv('customers-data-modif.csv')
data = dataset.iloc[:, [2, 3]].values
```

```python
x = np.asarray(data) # diubah formatnya menjadi array
print(x)
```

```python
# lakukan Hierachical Klastering dan plot dendogram
plt.figure(figsize=(25,10))
plt.title('Dendogram Hierarchical Klastering')
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
model.fit(data)
labels = model.labels_
print(labels)
```

```python
plt.scatter(data[labels==0, 0], data[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(data[labels==1, 0], data[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(data[labels==2, 0], data[labels==2, 1], s=50, marker='o', color='green')
#plt.scatter(data[labels==3, 0], data[labels==3, 1], s=50, marker='o', color='purple')
#plt.scatter(data[labels==4, 0], data[labels==4, 1], s=50, marker='o', color='orange')
plt.show()
```
